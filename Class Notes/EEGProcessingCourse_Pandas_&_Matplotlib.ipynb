# 2. Data Structures in Pandas: Series and DataFrames

##Series

```python
import pandas as pd


# Creating a Series for subject ages
subject_ages = pd.Series([29, 35, 42, 58], index=['Subject1', 'Subject2', 'Subject3', 'Subject4'])
print(subject_ages)

print(30*'-')

# Accessing a single element by index
print(subject_ages['Subject2'])  # Output: 35

print(30*'-')


# Accessing multiple elements
print(subject_ages[['Subject1', 'Subject3']])

print(30*'-')

# Performing arithmetic operations
subject_ages += 1
print(subject_ages)
```

#3. Data Exploration

## Inspecting Data with head(), tail(), info(), and describe()

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'Gender': ['F', 'M', 'M', 'M', 'F'],
    'Score': [85, 88, 90, 85, 87]
}

df = pd.DataFrame(data)


# Displaying the first few rows
print("First few rows of the DataFrame:")
print(df.head())

print(30*'-')

# Displaying the last few rows
print("Last few rows of the DataFrame:")
print(df.tail(3))

print(30*'-')

print("Summary of the DataFrame:")
print(df.info())
```

## Accessing Columns, Rows, and Cells with iloc and loc

### 1. Accessing Columns

```python
# Accessing a single column
print("Accessing the 'Name' column:")
print(df['Name'])

# Accessing multiple columns
print("Accessing the 'Name' and 'Score' columns:")
print(df[['Name', 'Score']])
```

### Accessing Rows by Integer Location with iloc

```python
# Accessing a single row by its integer index
print("Accessing the second row:")
print(df.iloc[1])

print(30*'-')

# Accessing multiple rows by their integer indices
print("Accessing the first and third rows:")
print(df.iloc[[0, 2]])
```

### Accessing Rows and Columns by Labels with loc

```python
# Accessing a single cell by its row and column labels
print("Accessing the cell at row 1 and column 'Age':")
print(df.loc[1, 'Age'])

print(30*'-')

# Accessing a row and specific columns by their labels
print("Accessing the second row with 'Name' and 'Score' columns:")
print(df.loc[1, ['Name', 'Score']])
```

# 5. Cleaning and Manipulating Data

## Handling Missing Values with isna() and fillna()

```python
import pandas as pd

# Sample DataFrame with missing values
data = {
    'patient_id': [1, 2, 3, 4],
    'age': [25, 30, None, 45],
    'gender': ['F', 'M', 'F', None],
    'condition': ['Healthy', 'Epileptic', 'Healthy', 'Healthy']
}
df = pd.DataFrame(data)


# Identifying missing values
print("Identifying missing values:")
print(df.isna())
```

```python
# Filling missing values with a specific value
df_filled = df.fillna({
    'age': df['age'].mean(),  # Fill missing ages with the mean age
    'gender': 'Unknown'       # Fill missing genders with 'Unknown'
})

print("Data after filling missing values:")
print(df_filled)

```

## Converting Data Types with astype()

```python
import pandas as pd
# Sample DataFrame with incorrect data types
data = {
    'patient_id': ['1', '2', '3', '4'],
    'age': ['25', '30', '35', '40'],
    'gender': ['F', 'M', 'F', 'M']
}
df = pd.DataFrame(data)

# Converting data types
df['patient_id'] = df['patient_id'].astype(int)
df['age'] = df['age'].astype(int)

print("Data after converting data types:")
print(df)

print(30*'-')

print("Data types:")
print(df.dtypes)
```

# 5. Data Transformation

## Applying Functions with apply(), map(), applymap()

```python
# Applying a function to a column
df['age_squared'] = df['age'].apply(lambda x: x ** 2)
print("Data after applying function with apply():")
print(df)
```

```python
# Mapping values in a column
df['gender_mapped'] = df['gender'].map({'F': 0, 'M': 1})
print("Data after mapping values with map():")
print(df)
```

## Iterating Over DataFrame Rows

```python
# Iterating over DataFrame rows
print("Iterating over DataFrame rows with iterrows():")
for index, row in df.iterrows():
    print(f"Index: {index}, Patient ID: {row['patient_id']}, Age: {row['age']}")
```

#2. Basic Plotting with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating a simple line plot
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

## Customizing Plots with Titles, Labels, and Legends

```python
# Example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating a customized plot
plt.plot(x, y1, label="channel 1")
plt.plot(x, y2, label="channel 2")
plt.title("Sine and Cosine Waves")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

## Types of Plots

### Line Plots

```python
# Example data
x = np.linspace(0, 10, 100)
y = np.exp(x)

# Creating a line plot
plt.plot(x, y, color='r', linestyle='--', marker='o')
plt.title("Exponential Growth")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

```

### Scatter plots

```python
# Example data
x = np.random.rand(100)
y = np.random.rand(100)

# Creating a scatter plot
plt.scatter(x, y, color='b', alpha=0.2)
plt.title("Random Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

```

### Bar Charts

```python
# Example data
categories = ['0-18', '18-25', '25-35', '35-50', '50-100']
values = [10, 20, 25, 21, 12]

# Creating a bar chart
plt.bar(categories, values, color='g')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### Histograms

```python
# Example data
data = np.random.randn(1000)

# Creating a histogram
plt.hist(data, bins=30, color='purple', alpha=0.1)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Subplots and Layouts

### Creating Multiple Plots in a Single Figure with subplot()

```python
# Example data
time = np.linspace(0, 10, 100)
eeg_signal1 = np.sin(2 * np.pi * 7 * time) + 0.5 * np.random.randn(100)
eeg_signal2 = np.cos(2 * np.pi * 7 * time) + 0.5 * np.random.randn(100)

# Creating subplots
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(time, eeg_signal1)
plt.title("Input EEG")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.subplot(1, 2, 2)
plt.plot(time, eeg_signal2)
plt.title("Preprocessed EEG")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```
